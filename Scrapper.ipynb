```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
HTML = requests.get("https://all.accor.com/de/country/hotels-deutschland-pde.shtml")
soup = BeautifulSoup(HTML.text, 'html.parser')
```

```python
## different regions

links = soup.find_all("a")
list_of_regions = []
for link in links:
    region_str = str(link.get('href'))
    if ("region" in region_str):
        list_of_regions.append(region_str)
    else:
        pass
```

```python
## list of hotels in different regions

hotels_list = []
for hotel in range(len(list_of_regions)):
    get_url = list_of_regions[hotel]
    HTML = requests.get(get_url)
    soup = BeautifulSoup(HTML.text,'html.parser')
    hotel_div = soup.find(id="paginator-list-core")
    Links = hotel_div.find_all('a')
    for i in Links:
       hotels_list.append(i.get('href'))
```

```python
## Remove City pages

for i in range(len(hotels_list)):
    try:
        if("city" in hotels_list[i]):
            hotels_list.remove(hotels_list[i])
    except:
        pass
```

```python
## remove index pages

for i in range(len(hotels_list)):
    try:
        if("index.de" not in hotels_list[i]):
            hotels_list.remove(hotels_list[i])
    except:
        pass
```

```python
## name, Address, Phone, URl to csv

with open('Data.csv', 'w', newline='') as file:
    for i in range(len(hotels_list)):
        link = hotels_list[i]
        hotel = requests.get(link)
        soup = BeautifulSoup(hotel.text, 'html.parser')
        location = soup.find(id='section-location')
        name = location.find('h3').text
        url = hotels_list[i]
        address = (location.find('p').text)
        phone = (location.find('a').text)
        
        # write the csv file
        writer = csv.writer(file)
        if i == 0:
            writer.writerow(["Name", "Address", "Phone", "URL"])
        writer.writerow([name, address, phone, url])
```
