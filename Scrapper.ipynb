```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
HTML = requests.get("https://all.accor.com/de/country/hotels-deutschland-pde.shtml")
soup = BeautifulSoup(HTML.text, 'html.parser')
```

```python
## different regions

links = soup.find_all("a")
list_of_regions = []
for link in links:
    region_str = str(link.get('href'))
    if ("region" in region_str):
        list_of_regions.append(region_str)
    else:
        pass
```

```python
## list of hotels in different regions

hotels_list = []
for hotel in range(len(list_of_regions)):
    get_url = list_of_regions[hotel]
    HTML = requests.get(get_url)
    soup = BeautifulSoup(HTML.text,'html.parser')
    hotel_div = soup.find(id="paginator-list-core")
    links = hotel_div.find_all('a')
    for link in links:
       hotels_list.append(link.get('href'))
```

```python
## remove the city and region links

def remove_links(*to_remove):
    list = []
    for i in range(len(hotels_list)):
        if (to_remove[0] not in hotels_list[i] and to_remove[1] not in hotels_list[i]):
            list.append(hotels_list[i])
    return list

hotels_list = remove_links("city", "index..")
```

```python
## name, Address, Phone, URl to csv

with open('Data.csv', 'w', newline='') as file:
    for hotel in range(len(hotels_list)):
        link = hotels_list[hotel]
        hotel_link = requests.get(link)
        soup = BeautifulSoup(hotel_link.text, 'html.parser')
        hotel_info = soup.find(id='section-location')
        name = hotel_info.find('h3').text
        url = hotels_list[hotel]
        address = (hotel_info.find('p').text)
        phone = (hotel_info.find('a').text)
        
        # write the csv file
        writer = csv.writer(file)
        if hotel == 0:
            writer.writerow(["Name", "Address", "Phone", "URL"])
        writer.writerow([name, address, phone, url])
```
